```python
cap =cv2.VideoCapture(0)
cap.set(3,500)
cap.set(4,500)
background = cv2.imread('background_image.jpg')
bars = cv2.namedWindow('bars')
def hello():
    pass

cv2.createTrackbar('UH', 'bars' ,110,180, hello)
cv2.createTrackbar('US', 'bars' ,255,255, hello)
cv2.createTrackbar('UV', 'bars' ,255,255, hello)
cv2.createTrackbar('LH', 'bars' ,68,180, hello)
cv2.createTrackbar('LS', 'bars' ,55,255, hello)
cv2.createTrackbar('LV', 'bars' ,54,255, hello)

while True:
    ret ,image   = cap.read()
    hsv  = cv2.cvtColor(image ,cv2.COLOR_BGR2HSV)
    
    UH = cv2.getTrackbarPos('UH' , 'bars')
    US = cv2.getTrackbarPos('US' , 'bars')
    UV = cv2.getTrackbarPos('UV' , 'bars')
    LH = cv2.getTrackbarPos('LH' , 'bars')
    LS = cv2.getTrackbarPos('LS' , 'bars')
    LV = cv2.getTrackbarPos('LV' , 'bars')
    
    kernel = np.ones((3,3))
    upper_hsv  = np.array([UH ,US,UV])
    lower_hsv  = np.array([LH ,LS,LV])
    
#     upper_hsv  = np.array([100 ,150,0])
#     lower_hsv  = np.array([140 ,255,255])
    
    
    mask = cv2.inRange(hsv ,lower_hsv ,upper_hsv)
    mask = cv2.medianBlur(mask ,1)
    mask_inv =  255-mask
    mask = cv2.dilate(mask ,kernel,3)

    b =image[:,:,0]
    g =image[:,:,1]
    r =image[:,:,2]
    b  =cv2.bitwise_and(mask_inv ,b)
    g  =cv2.bitwise_and(mask_inv ,g)
    r  =cv2.bitwise_and(mask_inv ,r)
    frame_inv = cv2.merge((b,g,r))
    
    b = background[:,:,0]
    g = background[:,:,1]
    r = background[:,:,2]
    b  =cv2.bitwise_and(b ,mask)
    g  =cv2.bitwise_and(g ,mask)
    r  =cv2.bitwise_and(r ,mask)
    blanket_area= cv2.merge((b,g,r))
    final = cv2.bitwise_or(frame_inv ,blanket_area)
    

    
    
    
    
    cv2.imshow('original',image)
    cv2.imshow('blanket',final)
    k =cv2.waitKey(5)
    if k==27:
        break
cap.release()
cv2.destroyAllWindows()
```
